```python
df_data.isna().sum().sort_values(ascending = False)
```

```python
def null_percentage(df):
    null = pd.DataFrame(columns=['category','percentage'])
    for col in df.columns:
        if df[col].isna().values.any():
            percentage = 100*df[col].isna().sum()/df.shape[0]
            null = null.append({'category' : col, 'percentage' : percentage}, ignore_index=True)
    return null 
```

```python
null_percentage(df_data).sort_values(by='percentage',ascending=False)
```
